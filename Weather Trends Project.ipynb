```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nyc=pd.read_csv('nyc temp.csv',index_col='year',parse_dates=True)
glob=pd.read_csv('global temp.csv',index_col='year',parse_dates=True)
nyc.drop(['city'],axis=True, inplace=True)
nyc.columns=['nyc_temp']
nyc.head()
```

```python
glob.head()
glob.columns=['global_temp']
glob.head()
```

```python
## Calucate 20 yrs of moving average for NYC

nyc['NYC_MA 20 years']=nyc.nyc_temp.rolling(20).mean()
ax1=nyc['NYC_MA 20 years'].plot(figsize=(15,5),c='red',grid=True)
ax1.set_ylabel("NYC Temperature")
ax1.set_xlabel("Years From 1750-2013")
```

```python
## Calucate 20 yrs of moving average for global
glob['Global_MA 20 years']=glob.global_temp.rolling(20).mean()
ax2=glob['Global_MA 20 years'].plot(figsize=(15,5),c='green',grid=True)
ax2.set_ylabel("Global Temperature")
ax2.set_xlabel("Years From 1750-2013")
```

```python
# Combine two datasets into single dataset
temp=pd.concat([nyc,glob],axis=1)
```

```python
temp.head(40)
```

```python
ax=temp[['NYC_MA 20 years','Global_MA 20 years']].plot(figsize=(17,7),
                                                       grid=True,title=("NYC & Global Temperature Comparison"))

ax.set_xlabel("Years from 1750-2013")
ax.set_ylabel("Temperatures")
```
